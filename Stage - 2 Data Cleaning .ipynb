```python
def tranform_text(text):
    
    #Convert into 
    text = text.lower()
    text = nltk.word_tokenize(text) 
    
    y = []
    for i in text:
        if i.isalnum():
            y.append(i)
            
            
    text = y[:]
    y.clear()
    
    custom_list = ['rt']
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)
    print(y)
    
    return " ".join(y)
```

```python
df['transformed_text'] = df['text'].apply(tranform_text)
```
